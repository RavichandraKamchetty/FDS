# Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var, hue = "sex")
plt.show()
```

```python
sns.pairplot(var, vars = ["total_bill", "tip"], hue = "sex", hue_order = ["Female", "Male"])
plt.show()
```

```python
sns.pairplot(var, hue = "sex", x_vars = ["total_bill", "tip"])
plt.show()
```

```python
sns.pairplot(var, hue = "sex", y_vars = ["total_bill", "tip"])
plt.show()
```

```python
sns.pairplot(var, hue = "sex", hue_order = ["Female", "Male"], kind = "kde", diag_kind = "hist") 
plt.show()

# kind: scatter, kde, hist, reg
```

```python
sns.pairplot(var, hue = "sex", hue_order = ["Female", "Male"], markers = ["o", ">"]) 
plt.show()

# kind: scatter, kde, hist, reg
```
