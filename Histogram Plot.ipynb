```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "penguins" dataset
var = sns.load_dataset("penguins")

# Set the style
sns.set(style="whitegrid")

# Create a density plot for the "bill_length_mm" column
sns.kdeplot(data=var, x="bill_length_mm", fill=True, common_norm=False)

# Add title and labels
plt.title("Density Plot of Bill Length")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Density")

# Show the plot
plt.show()
```
