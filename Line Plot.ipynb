```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = [1, 2, 3, 4, 5, 6, 7]
var_1 = [2, 3, 4, 1, 5, 6, 7]

sns.lineplot(x = var, y = var_1)
plt.show()
```

```python
var = [1, 2, 3, 4, 5, 6, 7]
var_1 = [2, 3, 4, 1, 5, 6, 7]

x_1 = pd.DataFrame({"var": var, "var_1": var_1})

sns.lineplot(x = "var", y = "var_1", data = x_1)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins").head(20)
y_1
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = y_1, hue = "sex")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y_1, hue = "sex",
                markers = ['o', '>'], style = "sex", palette = "viridis", dashes = False, legend = "full")


plt.grid()
plt.title("Penguins Dataset", fontsize = 15)
plt.show()



# size = 10
```
