# Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, hue = "time")
plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, palette = "Dark2", linewidth = 2, saturation = 10)
plt.show()
```

```python
sns.violinplot(x = "time", y = "total_bill", data = var, order = ["Dinner", "Lunch"], inner = "quartile")
plt.show()

# inner: quartile, point, stick, box
```

```python
sns.violinplot(x = "day", y = "total_bill", data = var, hue = "sex", split = True, scale = "width")
plt.show()

# scale: count, width, area
```

```python
sns.violinplot(x = var["total_bill"])
plt.show()
```

```python
sns.violinplot(y = var["total_bill"])
plt.show()
```
