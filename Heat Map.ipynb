# Heat Map

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = np.linspace(1, 10, 20).reshape(4, 5)
var
```

```python
y = {"fontsize": 15, "color": 'r'}

sns.heatmap(var, vmin = 0, vmax = 10, cmap = "PuOr", annot = True, annot_kws = y, linewidth = 3, 
                    linecolor = 'y', cbar = True, xticklabels = True, yticklabels = True)
plt.show()
```

```python
var1 = np.linspace(1, 10, 10).reshape(2, 5)
var1
```

```python
sns.heatmap(var1, vmin = 0, vmax = 10, cmap = "PuOr", annot = True)
plt.show()
```

```python
ar = np.array([["a0", "a1", "a2", "a3", "a4"],
                 ["b0", "b1", "b2", "b3", "b4"]])

sns.heatmap(var1, vmin = 0, vmax = 10, cmap = "PuOr", annot = ar, fmt = "s")
```
