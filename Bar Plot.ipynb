# Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var)
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
# ci means confidence interval
# n_boot means bootstrap samples used to compute confidence intervals

sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", 
                    order = order_1, hue_order = ["Female", "Male"], errorbar=('ci', 95), n_boot = 2)    
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]

sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", 
                    order = order_1, hue_order = ["Female", "Male"], orient = 'v', 
                    palette = "Accent", saturation = 100, errcolor = "c", errwidth = 3, capsize = 0.5)    
plt.show()
```
