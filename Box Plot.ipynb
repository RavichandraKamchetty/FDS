# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style = "whitegrid")

sns.boxplot(x = "day", y = "total_bill", data = var)

plt.show()
```

```python
sns.set(style = "whitegrid")

sns.boxplot(x = "day", y = "total_bill", data = var, hue = "time")

plt.show()
```

```python
sns.set(style = "whitegrid")

sns.boxplot(x = "day", y = "total_bill", data = var, hue = "sex", color = 'g', 
                                    order = ["Fri", "Sat", "Sun", "Thur"], showmeans = True)

plt.show()
```

```python
sns.set(style = "whitegrid")

sns.boxplot(x = "day", y = "total_bill", data = var, color = 'w', 
                    showmeans = True, meanprops = {"marker": "+", "markeredgecolor": "b"}, 
                       linewidth = 2, palette = "Accent")

plt.show()
```
